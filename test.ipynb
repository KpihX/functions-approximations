# 1. Test of affine approximation

```python
from func_approx import AffineApprox

fig, ax = set_fig("Linear Approximation")

x = [1, 2, 3, 4, 5]
y = [3, 6, 8, 11, 12]
ax.plot(x, y, 'o', label="")
    
affine_approx = AffineApprox(x, y)
print(f"affine_approx: {affine_approx}")
x0 = 2.5
print(f"affine_approx({x0}) = {affine_approx.predict(x0)}")
y0 = 10
print(f"affine_approx^(-1)({y0}) = {affine_approx.predict_y(y0)}")
print(f"corr = {affine_approx.corr}")
print(f"errors = {affine_approx.errors}")
print(f"moy_err = {affine_approx.moy_err}")
print(f"quad_err = {affine_approx.quad_err}")

x_plot = np.linspace(0, 10, 1000)
affine_approx.plot(ax, x_plot)
```

```python
from func_approx import ExpApprox

fig, ax = set_fig("Exponential Approximation")

x = [1, 2, 3, 4, 5]
y = [2, 5, 15, 45, 80]
ax.plot(x, y, 'o', label="")
    
exp_approx = ExpApprox(x, y)
print(f"exp_approx: {exp_approx}")
x0 = 2.5
print(f"exp_approx({x0}) = {exp_approx.predict(x0)}")
y0 = 10
print(f"exp_approx^(-1)({y0}) = {exp_approx.predict_y(y0)}")
print(f"corr = {exp_approx.corr}")
print(f"errors = {exp_approx.errors}")
print(f"moy_err = {exp_approx.moy_err}")
print(f"quad_err = {exp_approx.quad_err}")

x_plot = np.linspace(0, 5, 100)
exp_approx.plot(ax, x_plot)
```
