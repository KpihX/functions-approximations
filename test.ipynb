# Test of functions approximations

## 0. Prerequisites

```python
import numpy as np
from utils import set_fig
```

# 1. Test of affine approximation

```python
from func_approx import AffineApprox

fig, ax = set_fig("Linear Approximation")

x = [1, 2, 3, 4, 5]
y = [3, 6, 8, 11, 12]
ax.plot(x, y, 'o', label="")
    
affine_approx = AffineApprox(x, y)
print(f"affine_approx: {affine_approx}")
x0 = 2.5
print(f"affine_approx({x0}) = {affine_approx.predict(x0)}")
y0 = 10
print(f"affine_approx^(-1)({y0}) = {affine_approx.predict_y(y0)}")
print(f"corr = {affine_approx.corr}")
print(f"abs_errors = {affine_approx.abs_errors}")
pos_abs_errors = [abs(err) for err in affine_approx.abs_errors]
print(f"max_abs_error = {max(pos_abs_errors)}")
print(f"min_abs_error = {min(pos_abs_errors)}")
print(f"avg_abs_err = {affine_approx.avg_abs_err}")
print(f"quad_err = {affine_approx.avg_quad_err}")
print(f"rel_errors = {affine_approx.rel_errors}")
pos_rel_errors = [abs(rel_err) for rel_err in affine_approx.rel_errors]
print(f"max_rel_error = {max(pos_rel_errors)}")
print(f"min_rel_error = {min(pos_rel_errors)}")
print(f"avg_rel_err = {affine_approx.avg_rel_err}")
print(f"avg_rel_err2 = {affine_approx.avg_rel_err2}")

x_plot = np.linspace(0, 10, 1000)
affine_approx.plot(ax, x_plot)
```
